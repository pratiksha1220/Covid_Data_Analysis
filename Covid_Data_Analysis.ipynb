```python
import pandas as pd
```

```python
#Loading the dataset

covid = pd.read_csv('covid.csv') #On this line, I have stored my dataset in a variable called covid. 
```

```python
#Fetching first 5 records of the data

covid.head() 
```

```python
#fetching the number of rows and columns in the dataset

covid.shape
```

```python
#Fetching the names of columns in the dataset
covid.columns
```

```python
#Checking if the columns have null values

covid.isna().any() #True value in the output represents presence of null values and false values rpresents absence of null values 
```

```python
#Fetching the sum of null values in each column

covid.isna().sum()
```

```python
#Getting the cases in India

india_case=covid[covid["location"]=="India"] 
```

```python
#Fetching first 5 records of india_case

india_case.head()
```

```python
#Fetching last 5 columns of india_case

india_case.tail()
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#Making a dataframe for last 5 days cases in India

india_last_5_days=india_case.tail()
```

```python
#Total cases in last 5 days in India

sns.set(rc={'figure.figsize':(7,7)})
sns.lineplot(x="date",y="total_cases",data=india_last_5_days)
plt.show()
```

```python
#Total covid tests in last 5 days

sns.set(rc={'figure.figsize':(10,12)})
sns.lineplot(x="date",y="total_tests",data=india_last_5_days)
plt.show()
```

```python
#Creating a dataframe for Brazil Cases and fetching first 5 rows of it.

brazil_case=covid[covid["location"]=="Brazil"] 
brazil_case.head()
```

```python
brazil_case.tail()
```

```python
#creating a dataframe for brazil for last 5 days

brazil_last_5_days=brazil_case.tail()
```

```python
#Total cases in last 5 days

sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="total_cases",data=brazil_last_5_days)
plt.show()
```

```python
#Understanding cases of India, China and Japan

india_japan_china=covid[(covid["location"] =="India") | (covid["location"] =="China") | (covid["location"]=="Japan")]
```

```python
#Plotting growth of cases across China, India and Japan

sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="location",y="total_cases",data=india_japan_china,hue="date")
plt.show()
```

```python
#Getting latest data

last_day_cases=covid[covid["date"]=="2020-05-24"]
last_day_cases
```

```python
#Sorting data w.r.t total_cases

max_cases_country=last_day_cases.sort_values(by="total_cases",ascending=False)
max_cases_country
```

```python
#Top 5 countries with maximum cases

max_cases_country[1:6]
```

```python
#Making bar-plot for countries with top cases

sns.barplot(x="location",y="total_cases",data=max_cases_country[1:6],hue="location")
plt.show()
```
